```python
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
data = np.genfromtxt('KNMI_hourly_debilt_202311.txt',delimiter=',')
data_daily = np.genfromtxt('KNMI_daily_debilt_202311.txt',delimiter=',')
```

```python
temp = data[:,3]
day = data[:,1]
hour = data[:,2]

x = np.arange(len(temp))

temp_daily_max = data_daily[:,4]
day_daily = data_daily[:,1]
dates = [datetime.strptime(d, "%Y%m%d") for d in day_daily]
print(temp_daily_max)
print(dates)
```

```python
plt.figure(dpi=300)

plt.plot(x/24,temp/10)
plt.xlabel('t in days')
plt.ylabel('Temperature T ($^\circ$C)')
plt.title('Hourly temperature De Bilt November 2023')
plt.ylim(-3,17)



plt.figure(dpi=300)

plt.plot(np.arange(len(day_daily)),temp_daily_max/10)
plt.title('Max daily temperature De Bilt November 2023')
plt.ylabel('Temperature T ($^\circ$C)')
plt.xlabel('t in days')
plt.ylim(-3,17)

```

```python
fig, axes = plt.subplots(2, 1, figsize=(8, 6), dpi=300, sharex=False)

# --- First plot: hourly temperature ---
axes[0].plot(x/24, temp/10)
axes[0].set_xlabel("t in days")
axes[0].set_ylabel("Temperature T ($^\circ$C)")
axes[0].set_title("Hourly temperature De Bilt November 2023")
axes[0].set_ylim(-3, 17)
axes[0].set_xlim(0, 30)


# --- Second plot: daily max temperature ---
axes[1].plot(np.arange(len(day_daily)), temp_daily_max/10)
axes[1].set_xlabel("t in days")
axes[1].set_ylabel("Temperature T ($^\circ$C)")
axes[1].set_title("Max daily temperature De Bilt November 2023")
axes[1].set_xlim(0, 30)

# --- Layout tweaks ---
plt.tight_layout()
plt.show()
```
